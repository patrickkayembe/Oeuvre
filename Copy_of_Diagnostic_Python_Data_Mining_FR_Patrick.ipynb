### Q17. Mini-exercice de synthèse

On dispose de la liste suivante :
```python
donnees = [
    {'nom':'Amina', 'age':22, 'note':14},
    {'nom':'Brice', 'age':24, 'note':9},
    {'nom':'Chantal', 'age':23, 'note':16}
]
```
Écrivez un code qui :
1. transforme cette liste en `DataFrame`
2. ajoute une colonne `admis`
3. affiche uniquement les étudiants admis.

````python
# Écrivez votre réponse ici
import pandas as pd

donnees = [
    {'nom':'Amina', 'age':22, 'note':14},
    {'nom':'Brice', 'age':24, 'note':9},
    {'nom':'Chantal', 'age':23, 'note':16}
]


df = pd.DataFrame(donnees)

df['admis'] = df['note'] >= 10


admis = df[df['admis'] == True]

print(admis)
````
